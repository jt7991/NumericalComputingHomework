## Problem 2
### Part A


```
def power_method(A,x,tol,verbose=False):
  val_array = []
  
  y = np.dot(A,x)
  val = max(y.min(), y.max(),key=abs)
  vect = y/val
  val_array.append(val)
  
  y = np.dot(A,vect)
  val = max(y.min(), y.max(),key=abs)
  vect = y/val
  val_array.append(val)
  i = 1
 
  while(abs(val_array[i] - val_array[i-1])>tol):
    if(i>1000):
      print("Is not converging")
      return(val,vect)
    y = np.dot(A,vect)
    val =max(y.min(), y.max(),key=abs)
    vect = y/val
    if(verbose):
      print("Value:",val)
      print("Vector:",vect)
      print("A*b:",y)
    val_array.append(val)
    i+=1
  return val,vect
```

### Part B

```
val,vect = power_method(A, np.matrix('1;1;1'),10**-3)
print("i. Approximate Eigenvalue:",val)
print("   Approximate Eigenvector:\n",vect)
val2,vect2 = power_method(A, np.matrix('1;-1;1'),10**-3)
print("ii. Approximate Eigenvalue:",val2)
print("    Approximate Eigenvector:\n",vect2)
```
