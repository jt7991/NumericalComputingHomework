Question 6 a)

```
def my_mean(fun, a, b, N):
  if not a<=b:
    return
  h = (b-a)/N
  
  sum = 0
  for j in range(1,N):
    sum = sum + fun(a+(j-1)*h)
  
  return (sum/N)
    
```

b)

```
def my_fun(x):
  return (x*(math.e**x))
```

c)

```
N = np.array([10,20,40,80,160,320,640])
N_means=np.array([])
for i in N:
  N_means = np.append(N_means,my_mean(my_fun,-1,1,i))
N_means_err = np.array([abs(1/math.e -x) for x in N_means ])

plt.loglog(N,N_means_err)
plt.xlabel("N")
plt.ylabel("Error")
print("\n\n\n\n\n\n")
plt.show()

print("\n\n\n\n\n\n\n\n\n\n\n\n\n")
```

d)

```
print("     N        |     Approximation      |   Error")
for i in range(0,len(N)):
  print("{}              {}      {}\n".format(N[i],N_means[i],N_means_err[i]))
```

The larger N becomes, the error gets smaller and the approximation gets closer to the actual value.
